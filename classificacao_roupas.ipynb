```python
import tensorflow
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados = keras.datasets.fashion_mnist
((x_train, y_train), (x_test, y_test)) = dados.load_data()
```

```python
# Labels
labels = ['T-shirt/top','Trouser','Pullover',	'Dress',	'Coat',	'Sandal',	'Shirt',	'Sneaker','Bag','Ankle boot']	
```

```python
for imagem in range(10):
  plt.subplot(2,5,imagem+1)
  plt.imshow(x_train[imagem])
  plt.title(labels[y_train[imagem]])
```

```python
plt.imshow(x_train[0])
plt.colorbar()
```

```python
# normalização
x_train = x_train/float(255)
```

```python
model = keras.Sequential([
                    keras.layers.Flatten(input_shape=(28,28)),
                    keras.layers.Dense(256,activation=tensorflow.nn.relu),
                    keras.layers.Dropout(0.2),
                    keras.layers.Dense(10, tensorflow.nn.softmax)
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python

historico = model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
plt.plot(historico.history['acc'])
plt.plot(historico.history['val_acc'])
plt.title('Acurácia por Épocas')
plt.xlabel('Épocas')
plt.ylabel("Acurácia")
plt.legend(['Treino', 'Validação'])
```

```python

plt.plot(historico.history['loss'])
plt.plot(historico.history['val_loss'])
plt.title('Perda por Épocas')
plt.xlabel('Épocas')
plt.ylabel("Perda")
plt.legend(['Treino', 'Validação'])
```

```python
perda, acuracia = model.evaluate(x_test, y_test)
```

```python

print('perda do teste: ', perda)
print('Acurácia do teste: ', acuracia)
```
